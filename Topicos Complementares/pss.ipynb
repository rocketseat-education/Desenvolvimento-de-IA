```python
import pandas as pd
import seaborn as sns
import ppscore as pps
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
```

### Carregar e visualizar os dados - Regressão

```python
# Carregar os dados - Regressão
df = pd.read_csv("datasets/colesterol.csv")
```

```python
# Visualizar estrutura dos dados
df.info()
```

```python
df.describe()
```

```python
df.head(10)
```

```python
# Ajustar Dataframe
df.drop(columns=["Id"], axis=1, inplace=True)

# Aplicar One Hot Encoding nas variaveis categóricas
df = pd.get_dummies(df, columns=["Grupo Sanguíneo", "Fumante", "Nível de Atividade"])
```

```python
# Dataframe atualizado
df.head(10)
```

### Calcular PPS - Regressão

```python
# Calcular PPS entre as variáveis
pps_matrix_reg = pps.matrix(df)
pps_matrix_reg
```

```python
# Ajustar a matriz para fazer o Plot
pps_matrix_reg_pivo = pps_matrix_reg[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
pps_matrix_reg_pivo
```

```python
# Plotar a matriz de PPS
plt.figure(figsize=(15, 8))
sns.heatmap(pps_matrix_reg_pivo, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Predictive Power Score (PPS) - Modelo Colesterol")
plt.show()
```

```python
# Calcular o PPS entre duas variáveis
pps.score(df, "Peso", "Colesterol")
```

#### Matriz de Correlação - Regressão

```python
# Calcular a matriz de correlação
correlation_matrix_reg = df.corr()
correlation_matrix_reg
```

```python
# Plotar a matriz de PPS
plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix_reg, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação - Modelo Colesterol")
plt.show()
```

### Treinar Modelo Linear - Regressão

```python
# Separar X e y
X = df.drop(columns=["Colesterol"])
y = df["Colesterol"]
```

```python
model_reg = LinearRegression()
model_reg.fit(X, y)
```

```python
# Avaliar a importância das features com base nos coeficientes do modelo
feat_importance_reg = pd.Series(model_reg.coef_, index=X.columns).sort_values(ascending=True)
feat_importance_reg.plot(kind='barh', figsize=(10, 6))

plt.xlabel("Importância")
plt.ylabel("Features")
plt.title("Importância das Features - Regressão")
plt.show()
```

### Carregar e preprar dados - Classificação

```python
df2 = pd.read_csv("datasets/fruit_quality.csv")
```

```python
# Visualizar estrutura dos dados
df2.info()
```

```python
df.describe()
```

```python
# Visualizar Df
df2.head(10)
```

```python
# Ajustar Dataframe
# remover a coluna A_id
df2.drop(columns=["A_id"], axis=1, inplace=True)

# Transforma a variável Quality
df2['Quality'] = (df2['Quality'] == 'good').astype(int)
```

```python
df2.head(10)
```

### Calcular PPS - Classificação

```python
# Calcular PPS entre as variáveis
pps_matrix_class = pps.matrix(df2)
pps_matrix_class
```

```python
# Ajustar a matriz para fazer o Plot
pps_matrix_class_pivo = pps_matrix_class[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
pps_matrix_class_pivo
```

```python
# Plotar a matriz de PPS
plt.figure(figsize=(15, 8))
sns.heatmap(pps_matrix_class_pivo, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Predictive Power Score (PPS) - Modelo Qualidade de Frutas")
plt.show()
```

```python
# Calcular o PPS entre duas variáveis
pps.score(df2, "Size", "Quality")
```

#### Matriz de Correlação - Classificação

```python
# Calcular a matriz de correlação
correlation_matrix_class = df2.corr()
correlation_matrix_class
```

```python
# Plotar a matriz de PPS
plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix_class, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação - Modelo Qualidade de Frutas")
plt.show()
```

### Treinar Modelo Linear - Classificação

```python
# Separar X e y
X = df2.drop(columns=["Quality"])
y = df2["Quality"]
```

```python
model_lr = LogisticRegression()
model_lr.fit(X, y)
```

```python
# Avaliar a importância das features com base nos coeficientes do modelo
feat_importance_class = pd.Series(model_lr.coef_[0], index=X.columns).sort_values(ascending=True)
feat_importance_class.plot(kind='barh', figsize=(10, 6))

plt.xlabel("Importância")
plt.ylabel("Features")
plt.title("Importância das Features - Classificação")
plt.show()
```
